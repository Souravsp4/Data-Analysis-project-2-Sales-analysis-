#### Import Necessary Libraries

```python
import pandas as pd
import os 
```

#### Merging 12 months of sales data into single file

```python
df = pd.read_csv("./SalesAnalysis/Sales_Data/Sales_April_2019.csv")
files = [file for file in os.listdir('./SalesAnalysis/Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./SalesAnalysis/Sales_Data/"+file)
    all_months_data= pd.concat([all_months_data,df])

all_months_data.head()

all_months_data.to_csv("all_data.csv",index=False)

```

#### Read in updated dataframe

```python
all_data=pd.read_csv("all_data.csv")
all_data.head()
```

#### Drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2]!="Or"]
```

#### Convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])# make float
```

#### Augment data with additional column

#### Task2: Add month column

```python
all_data['Month'] =all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Task 3 : Add a sales column

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
all_data=all_data.drop(['Quantity Oedered'], axis=1)
```

```python
all_data.head()
```

#### Task 4 : Add a city column

```python
# let's use .apply()
def get_city(adress):
    return adress.split(',')[1]

def get_state(adress):
    return adress.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

#### Q.1) What was the best month for sales ? How much was earned 

```python
results=all_data.groupby('Month')[['Quantity Ordered', 'Price Each', 'Sales']].sum()
results.head()
```

#### Plotting

```python
import matplotlib.pyplot as plt
months =range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month number')
plt.show()
```

#### Q.2)  What city had the highest no of sales ?

```python
results=all_data.groupby('City')[['Quantity Ordered', 'Price Each', 'Sales']].sum()
results
```

#### Plotting

```python
import matplotlib.pyplot as plt
# city_sales = all_data.groupby('City')['Sales'].sum().reset_index()
cities = [city for city , df in all_data.groupby('City')]

# Create a bar plot
plt.bar(cities, results['Sales'])

plt.ylabel('Sales in USD($)')
plt.xlabel('City name')
plt.xticks(rotation='vertical')
plt.show()
```

#### Q.3) What time should we display advertisements to maximize the likelihood of purchases ?

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
all_data.groupby(['Hour']).count()
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.grid()
plt.show()

# My reccomendation is arround 11am (11) or 7pm(19)
```

#### Q.4) what products are most solved often together ? 

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list= row.split(',')
    count.update(Counter(combinations(row_list,2)))

count.most_common(10)
```
